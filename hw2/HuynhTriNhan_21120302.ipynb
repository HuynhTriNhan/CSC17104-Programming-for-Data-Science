## Tiền xử lý

```python
new_cols = {}
for col_name, col_vals in cols.items():
    if '/' in    col_name:
        new_values = []
        for value in col_vals:
            new_values.append(float(value))
        new_cols[col_name] = new_values
    
    elif col_name in ['Started on', 'Completed']:
        new_values = []
        for value in col_vals:
            
            new_values.append(dt.datetime.strptime(value, "%d/%m/%Y %H:%M"))
        new_cols[col_name] = new_values
    
    elif col_name == 'Time taken':
        new_values = []
        for value in col_vals:
            h = 0;  m=0; s=0
            new_value = value.split(' ')

            for i in range(len(new_value)):
                if new_value[i] in ['hours','hour'] :
                    h = int(new_value[i-1])
                elif 'mins' in new_value[i]:
                    m = int(new_value[i-1])
                elif 'secs' in new_value[i]:
                    s = int(new_value[i-1])
            new_values.append(h*60 + m + s/60)
        new_cols[col_name] = new_values
    else:
        new_values = []
        for value in col_vals:
            new_values.append(value)
        new_cols[col_name] = new_values

# new_cols
# Kiểm tra kiểu dữ liệu các cột
for col in new_cols:
    print(col, type(new_cols[col][0]))
```

## Tiền xử lý + phân tích dữ liệu để trả lời cho câu hỏi 1

*Câu hỏi 1: Cột điểm tổng “Grade/10.00” có phân bố các giá trị như thế nào?*

Ở đây không cần tiền xử lý mà sẽ đi luôn vào phân tích dữ liệu.

Đầu tiên, ta sẽ xem thử median của cột "Grade/16.00".

```python
   
def check_bin(value, bounds: tuple):
    left_value, right_value = bounds
    if left_value == 0:
        if value >= left_value and value <= right_value:
            return True
    else: 
        if value > left_value and value <= right_value:
            return True
    return False

bin_grades={}
bins = [(0,4),(4,8),(8,12),(12,16)]   
for bin_range in bins:
    bin_grades[bin_range] = 0
 
for score in cols['Grade/16.00']:
    score = float(score)
    if check_bin(score,bins[0]):
        bin_grades[bins[0]] +=1
    elif check_bin(score,bins[1]):    
        bin_grades[bins[1]] +=1
    elif check_bin(score,bins[2]):
        bin_grades[bins[2]] +=1
    elif check_bin(score,bins[3]):
        bin_grades[bins[3]] += 1

print(bin_grades)
print(f'So hoc sinh co diem nam trong khoang [0,4] la {bin_grades[(0,4)]}')
print(f'So hoc sinh co diem nam trong khoang (4,8] la {bin_grades[(4,8)]}')
print(f'So hoc sinh co diem nam trong khoang (8,12] la {bin_grades[(8,12)]}')
print(f'So hoc sinh co diem nam trong khoang (12,16] la {bin_grades[(12,16)]}')

```

Để cảm thụ rõ hơn về phân bố các giá trị của cột "Grade/16.00", ta sẽ chia miền giá trị của cột này thành các khoảng (bin) và xem thử mỗi khoảng có bao nhiêu sinh viên.

Giả sử ta sẽ chia đoạn [0, 16] thành 4 bin như sau: [0, 4], (4, 8], (8, 12], (12, 16]. Ta có thể lưu kết quả vào một dictionary, trong đó: key là một tuple cho biết 2 đầu của một bin (thay tuple bằng list thì có được không?), value là số lượng sinh viên rơi vào bin này.



## Tiền xử lý + phân tích dữ liệu để trả lời cho câu hỏi 2

*Câu hỏi 2: Câu quiz nào có nhiều bạn làm sai nhất, câu nào có nhiều bạn làm sai nhì, ...?*

Ta có thể lưu kết quả vào một dictionary, trong đó: key là tên của cột ứng với câu quiz, value là điểm trung bình của cột. Dictionary này được sort tăng dần theo value.



```python
# Thống kê số điểm 0 trong từng câu hỏi
average_points = {}
for col_name, col_vals in cols.items():
    if '/1.00' in col_name:
        sum = 0
        for value in cols[col_name]:
            sum += float(value)
        average_points[col_name] = sum*1.0/len(cols[col_name])

print(average_points)
# Diem sau khi sort
sorted_average_points = dict(sorted(average_points.items(), key=lambda item: item[1]))
print(sorted_average_points)


print(f'Cau hoi co nhieu ban tra loi sai nhat tuong ung diem trung binh thap nhat la {list(sorted_average_points.keys())[0]} voi diem trung binh la {list(sorted_average_points.values())[0]}')
print(f'Cau hoi co nhieu ban tra loi sai nhi tuong ung diem trung binh thap thu nhi la {list(sorted_average_points.keys())[1]} voi diem trung binh la {list(sorted_average_points.values())[1]}')
                
 

```
