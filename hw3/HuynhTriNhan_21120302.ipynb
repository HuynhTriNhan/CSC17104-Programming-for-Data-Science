## Exercises 1

Use `re.search` to find whether a string contains a phone number. The pattern that you write should detect a phone number in the following strings.  
```
"Call me at 382-384-3840."  
"my number is (510) 849-3519. Call me!"
```  
And not find a match in the following strings. 
```
"my number is 510-849-35192"  
"here’s my number: 510-849.3519"
``` 
Consider making your own tests as well  

````python
testcase = [
    "Call me at 382-384-3840.",  
    "my number is (510) 849-3519. Call me!",
    "my number is 510-849-35192",
    "here’s my number: 510-849.3519"
]
regex = r'(\(?\d{3}\)?[-\s]?\d{3}[-\s]?\d{4})[\D]'
for test in testcase:
    match = re.search(regex, test)
    if match:
        print(match.group(1))
````

## Exercise 2

Use `re.sub` to alter the string below so that the dates have a common format that uses a dash for the day, month, and year separator.  
```
03/12/2018, 03.13.18, 03/14/2018, 03:15:2018
```

````python
testcase =[
    '03/12/2018',
    '03.13.18',
    '03/14/2018',
    '03:15:2018'
]
# Thay các ký tự / . : bằng dấu -
regex = r"(\d{2})[-:\.\/](\d{2})[-:\.\/](\d{2,4})"
for test in testcase:
    # Thay thế
    print(re.sub(regex, r"\1-\2-\3", test))
````

## Exercise 3

Consider the first five sentences of the novel “Little Women” below. Extract the spoken dialog from each sentence.

````python
text = '''
"Christmas won't be Christmas without any presents," grumbled Jo, lying on the rug.
"It's so dreadful to be poor!" sighed Meg, looking down at her old dress.
"I don't think it's fair for some girls to have plenty of pretty things, and other girls nothing at all," added little Amy, with an injured sniff.
"We've got Father and Mother, and each other," said Beth contentedly from her corner.
The four young faces on which the firelight shone brightened at the cheerful words, but darkened again as Jo said sadly, "We haven't got Father, and shall not have him for a long time."
'''
print(text)
````

````python
# Tách mỗi câu bởi những dấu kết thúc câu
sentences = re.split(r'[\.\?!]', text)
for sentence in sentences:
    # Loại bỏ khoảng trắng ở đầu và cuối câu và ký tụ \n
    sentence = sentence.strip()
    print(sentence)
````

## Exercise 4

In this exercise, you you working with ```email_test.txt``` file (attached), using Regular Expression. 
1. Extract name and email address of sender  
Output should be something like this
```
Name:
"MR. JAMES NGOLA." 
"Mr. Ben Suleman" 
"PRINCE OBONG ELEME" 
"PRINCE OBONG ELEME" 
"Maryam Abacha" 
```
```
Email address:
"james_ngola2002@maktoob.com"
"bensul2004nng@spinfinder.com"
"obong_715@epatra.com"
"obong_715@epatra.com"
"m_abacha03@www.com"
```
2. Find SUBJECT of All email from 01/06/2003 to 13/05/2004 


````python
import re
import datetime as dt
# 1. Extract name and email address of sender
# 2. Find SUBJECT of All email from 01/06/2003 to 13/05/2004

fh = open(r"Data/email_test.txt", mode="r", encoding="utf-8").read()
# Số letter = 1331
# Tìm kiếm được date name address not null: 849
# date name address not null: 981
# date not null: 1051

regex_from = r"From: (.*) <(.*)>"
regex_date = r"Date: \w+, (\d{1,2}) (\w{3}) (\d{4}) (\d{2}):(\d{2}):(\d{2})"
regex_subject = r"Subject: (.*)"

# Chia nhỏ các email
letters = re.split(r"From r ", fh)
# print(letters)
data_cleaned =[]

for letter in letters:
    match_from = re.search(regex_from, letter)
    if match_from:
        name = match_from.group(1)
        # Cắt bỏ dấu "" thừa
        if name[0] == '"':
            name = name[1:-1]
        address = match_from.group(2)
    else:
        name = None
        address = None
        
    # Chuẩn hóa ngày tháng năm
    match_date = re.search(regex_date, letter)
    if match_date:
        day = match_date.group(1)
        month = match_date.group(2)
        year = match_date.group(3)
        hour = match_date.group(4)
        minute = match_date.group(5)
        second = match_date.group(6)
        date = (dt.datetime(int(year), dt.datetime.strptime(month, '%b').month, int(day), int(hour), int(minute), int(second)))
    else:
        date = None
    
    match_subject = re.search(regex_subject, letter)
    if match_subject:
        subject = match_subject.group(1)
    else:
        subject = None  
         
    if name and address and date and subject:
        data_cleaned.append({
            "name": name,
            "address": address,
            "date": date,
            "subject": subject
        })
        


# In ra các email từ 01/06/2003 đến 13/05/2004

data_filtered = []
for letter in data_cleaned:
    if letter["date"] >= dt.datetime(2003, 6, 1) and letter["date"] <= dt.datetime(2004, 5, 13):
        data_filtered.append(letter)

print(f"Tìm thấy {len(data_cleaned)} thông tin đầy đủ (có ngày,tên người gửi, địa chỉ, subject) trong {len(letters)} email")
print("Thông tin 5 người gửi email tìm thấy đầu tiên")
print("\n\nName: ")
for letter in data_cleaned[:5]:
    print(letter["name"])
    
print("\nAddress: ")
for letter in data_filtered[:5]:
    print(letter["address"])
    
print(f"\nCác subject của các email từ 01/06/2003 đến 13/05/2004: Có: {len(data_filtered)} email")
for letter in data_filtered:
    print(letter["subject"])
        
        
    
````
